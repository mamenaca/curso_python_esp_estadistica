```python
import numpy as np
```

##Ejemplo 1: Gráfica simple

```python
x=np.arange(1,10+1)
y=x**2
```

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y)

#CAracterísticas adicionales

#Mostrar
plt.show()
```

##Ejemplo dos: Dos gráficas

```python
y1=25*np.exp(-x**2)
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y)
plt.plot(x,y1)

#CAracterísticas adicionales

#Mostrar
plt.show()
```

##Ej 3: configuraciones mínimas

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y,label="$x^2$")
plt.plot(x,y1,label="$25 * exp(x^2)$")

#Características adicionales
plt.grid()
plt.legend(loc="best")
#nombres de ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)


#Mostrar
plt.show()
```

##Ej anterior con más colores

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y,label="$x^2$",color=(1,0,0),alpha=0.2)
plt.plot(x,y1,label="$25 * exp(x^2)$",color="magenta")

#Características adicionales
plt.grid()
plt.legend(loc="best")
#nombres de ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)


#Mostrar
plt.show()
```

##Marcadores

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y,marker="$x$",label="$x^2$",color=(1,0,0),alpha=0.2)
plt.plot(x,y1,marker="$Y$",label="$25 * exp(x^2)$")

#Características adicionales
plt.grid()
plt.legend(loc="best")
#nombres de ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)


#Mostrar
plt.show()
```
