##Ejemplo 1: Gráfica simple

```python
x=np.arange(1,10+1)
y=x**2
```

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y)

#CAracterísticas adicionales

#Mostrar
plt.show()
```

##Ejemplo dos: Dos gráficas

```python
y1=25*np.exp(-x**2)
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y)
plt.plot(x,y1)

#CAracterísticas adicionales

#Mostrar
plt.show()
```

##Ej 3: configuraciones mínimas

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y,label="$x^2$")
plt.plot(x,y1,label="$25 * exp(x^2)$")

#Características adicionales
plt.grid()
plt.legend(loc="best")
#nombres de ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)


#Mostrar
plt.show()
```

##Ej anterior con más colores

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y,label="$x^2$",color=(1,0,0),alpha=0.2)
plt.plot(x,y1,label="$25 * exp(x^2)$",color="magenta")

#Características adicionales
plt.grid()
plt.legend(loc="best")
#nombres de ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)


#Mostrar
plt.show()
```

##Marcadores

```python
#crear el lienzo
plt.figure()

#Definir los gráficos
plt.plot(x,y,marker="$x$",label="$x^2$",color=(1,0,0),alpha=0.2)
plt.plot(x,y1,marker="$Y$",label="$25 * exp(x^2)$")

#Características adicionales
plt.grid()
plt.legend(loc="best")
#nombres de ejes
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)


#Mostrar
plt.show()
```

#Gráficas de área:

```python
plt.figure()#creacion del lienzo

#DEfinición de gráfica
plt.fill_between(x,y,y1,color="olive", alpha=.5) 
plt.plot(x,y,color="k")
plt.plot(x,y1,color="k")

#parámetros adiciones
plt.grid()

plt.xlabel("X")
plt.xlabel("Y")

plt.legend(loc="best")

plt.show()# Mostrar
```

###Uso del parámetro where=

```python
x=np.linspace(0,10,100)
y=x**2
y1=25*np.exp(-x**2)
filtro=x>3
plt.figure()#creacion del lienzo

#DEfinición de gráfica
plt.fill_between(x,y,y1,color=(1,0,0), alpha=.5,where=filtro) 
plt.plot(x,y,color="k")
plt.plot(x,y1,color="k")

#parámetros adiciones
plt.grid()

plt.xlabel("X")
plt.xlabel("Y")

plt.legend(loc="best")

plt.show()# Mostrar
```

###Ej where, con dos filtros

```python
filtro1=(x>0) & (x<=2.5)
filtro2=(x>2.5) & (x<=6)
filtro3=x>6
plt.figure()#creacion del lienzo

#DEfinición de gráfica
plt.fill_between(x,y,y1,color=(1,1,0), alpha=.5,where=filtro1) 
plt.fill_between(x,y,y1,color=(0,0,1), alpha=.5,where=filtro2) 
plt.fill_between(x,y,y1,color=(1,0,0), alpha=.5,where=filtro3) 
plt.plot(x,y,color="k")
plt.plot(x,y1,color="k")

#parámetros adiciones
plt.grid()

plt.xlabel("X")
plt.xlabel("Y")

plt.legend(loc="best")

plt.show()# Mostrar
```

###Texto en gráfica

```python
filtro1=(x>0) & (x<=2.5)
filtro2=(x>2.5) & (x<=6)
filtro3=x>6
plt.figure()#creacion del lienzo

#DEfinición de gráfica
plt.fill_between(x,y,y1,color=(1,1,0), alpha=.5,where=filtro1) 
plt.fill_between(x,y,y1,color=(0,0,1), alpha=.5,where=filtro2) 
plt.fill_between(x,y,y1,color=(1,0,0), alpha=.5,where=filtro3) 
plt.plot(x,y,color="k")
plt.plot(x,y1,color="k")

#parámetros adiciones
plt.text(1.6,30,"Intercepto",ha="center",va="center",fontsize=10)
plt.grid()

plt.xlabel("X")
plt.xlabel("Y")

plt.legend(loc="best")

plt.show()# Mostrar
```

###Flechas en gráfica:

```python
plt.figure()#creacion del lienzo

#DEfinición de gráfica
plt.fill_between(x,y,y1,color=(1,1,0), alpha=.5,where=filtro1) 
plt.fill_between(x,y,y1,color=(0,0,1), alpha=.5,where=filtro2) 
plt.fill_between(x,y,y1,color=(1,0,0), alpha=.5,where=filtro3) 
plt.plot(x,y,color="k")
plt.plot(x,y1,color="k")

#parámetros adiciones
#plt.text(1.6,30,"Intercepto",ha="center",va="center",fontsize=10)
plt.grid()

plt.gca().annotate(
    "Intercepto",
    xy=(1.6,5),
    xytext=(2,50),
    va="center",
    ha="center",
    arrowprops=dict(facecolor="black",shrink=0.05)


)

plt.xlabel("X")
plt.xlabel("Y")

plt.legend(loc="best")

plt.show()# Mostrar
```

###Exportar gráficas

Es posible exportar gráficas vectorizadas (modificables) en pdf

```python
plt.figure()#creacion del lienzo

#DEfinición de gráfica
plt.fill_between(x,y,y1,color=(1,1,0), alpha=.5,where=filtro1) 
plt.fill_between(x,y,y1,color=(0,0,1), alpha=.5,where=filtro2) 
plt.fill_between(x,y,y1,color=(1,0,0), alpha=.5,where=filtro3) 
plt.plot(x,y,color="k")
plt.plot(x,y1,color="k")

#parámetros adiciones
#plt.text(1.6,30,"Intercepto",ha="center",va="center",fontsize=10)
plt.grid()

plt.gca().annotate(
    "Intercepto",
    xy=(1.6,5),
    xytext=(2,50),
    va="center",
    ha="center",
    arrowprops=dict(facecolor="black",shrink=0.05)


)

plt.xlabel("X")
plt.ylabel("Y")

plt.legend(loc="best")

plt.savefig("grafica.pdf")# guarda en la carpeta de ejecución
plt.show()# Mostrar

```
