##Barras múltiples

```python
                  AXES0,AXES1..AXES2
```

```python
def poner_etiqueta(edad):
  if 0<edad <12:
    return "< 12"
  elif 12<=edad<18:
    return "[12,18)"
  elif 18<=edad <40:
    return "[18,40)"
  elif edad>=40:
    return "viejos"
  else:
    return "Error de edad"
```

```python
edad_etiq=titanic.Age.apply(poner_etiqueta)
titanic["edad_etiq"]=edad_etiq
titanic.head()
```

```python
titanic.groupby(["edad_etiq","Pclass"]).Sex.value_counts().unstack(1).plot.bar()
```

```python
titanic.groupby(["edad_etiq"]).Sex.value_counts().unstack(1).plot.bar()
```

Más avanzado de torta:

```python
plt.figure()

titanic.Pclass.value_counts().plot.pie(
    autopct="%.2f",
    labels=["Primera clase","segunda clase","tercera clase"],
    colors=["crimson","gold","skyblue"],
    title="proporción de clases en el titanic"
    )

plt.show()
```

zip: comprimir

```python
conteo_clases=titanic.Pclass.value_counts().sort_index(ascending=True)
conteo_clases
```

```python
etiquetas=[]
for ind,val in zip(conteo_clases.index,conteo_clases.values):
  etiquetas.append("Clase "+ str(ind)+"\n"  +str(val) )
etiquetas
```

```python
plt.figure()

titanic.Pclass.value_counts().plot.pie(
    autopct="%.2f",
    labels=etiquetas,
    colors=["crimson","gold","skyblue"],
    title="proporción de clases en el titanic"
    )

plt.show()
```
