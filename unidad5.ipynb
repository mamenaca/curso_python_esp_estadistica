##¿Cuál es la película con mayor y menor duración?

```python
filtro_movie=netflix.type=="Movie"
funcion=lambda x: int(x.replace(" min",""))
maxi=netflix[filtro_movie].duration.apply(   funcion  ).max()
mini=netflix[filtro_movie].duration.apply(   funcion  ).min()
```

```python
maxi,mini
```

```python
filtroM=netflix[filtro_movie].duration.apply(   funcion  ) ==maxi
filtrom=netflix[filtro_movie].duration.apply(   funcion  ) ==mini
```

```python
sum(filtrom)
```

```python
netflix[filtro_movie][filtroM].title.values[0]
```

```python
netflix[filtro_movie][filtrom].title.values[0]
```
