##Diferencia de conjuntos: Qué tiene el conjunto 1 que no tiene el conjunto 2

```python
comida={"platano", "hamburguesa","sancocho","manzana","pollo frito","fresa"}
frutas={"limon","pera","platano","manzana"}
comida-frutas
```

```python
frutas-comida
```

```python
frutas.difference(comida)
```

```python
frutas
```

```python
comida
```

Es decir, lo que se puede quitar(restar), se quita, pero ninguno de los conjuntos se modificó. Es decir, la operación no es In-place

##Diferencia de conjuntos con update: Se actualiza la resta y modifica al primer conjunto

```python
frutas.difference_update(comida)
```

```python
frutas
```

##Intersección de conjuntos, también existe con la forma updated

```python
frutas.intersection_update(comida)
```
