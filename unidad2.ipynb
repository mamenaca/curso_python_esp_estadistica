##Parámetros adicionales: **kwargs

*   Como son parámetros adicionales, debe dárseles un nombre y valor
*   Internamente se interpretan entonces los **kwargs como un diccionario (si se imprimen los **kwargs devuelve un diccionario)
*   Dentro de una función permite llamar a otra función, y se propagan los parámetros, es decir, se usan en funciones "más internas"
*   Una función puede tener cero argumentos posicionales, solo **kwargs

```python
def colores(**kwargs):
  print(kwargs)
```

```python
colores(a=5,rojo="stef")
```

```python
colores(10)
```

```python
import math
def area_cuadrado(a):
  return a*a
def area_trian(b,a):
  return b*a/2
def area_circ(r):
  return math.pi*r**2
```

```python
def area_figura(nombre,**parametros_caract):
  if nombre=="cuadrado":
    return area_cuadrado(**parametros_caract)
  elif nombre=="triangulo":
    return area_trian(**parametros_caract)
  elif nombre=="circulo":
    return area_circ(**parametros_caract)
```

```python
area_figura("cuadrado",a=5)
```

```python
area_figura("triangulo",a=4,b=3)
```

```python
area_figura("circulo",r=5)
```

¡Ojo! los **kwargs se deben recibir con el nombre de los argumentos que recibe la función objetivo

###Funciones que reciben funciones como arguments: Mapper, en python el mapper in-built es map()

La siguiente función aplica la función potencia con exp=2 por defecto

```python
def potencia_lista(funcion,lista):
  salida=[]
  for i in lista:
    salida.append(funcion(i))
  return salida
```

```python
jj=[1,2,6,23,88,6]
potencia_lista(potencia,jj)
```

¿Cómo modificar el código para aplicar la potencia a cualquier expotente?

```python
list(map(potencia,jj))
```

```python
potencia_lista(math.sqrt,jj)
```
