##Dos maneras de copiar variables en python: Shallow and deep (superficial y profunda)

###Shallow copy: Para elementos que no tengan niveles (ni lista de listas, ni tupla de tuplas, ni diccionarios)

```python
lista2=lista1.copy()
print(id(lista1),id(lista2))
```

```python
lista1.extend([444]) #recordar que el argumento es un iterable
lista1
```

```python
lista2
```

###Deep copy: Para elementos que **tengan** niveles (como lista de listas, tupla de tuplas o diccionarios)

###*Una lista es Mutable, pero puede tener elementos inmutables

```python
lista3=[2,[55,765,4],44,6,3,(3,44),55,6]
lista4=lista3.copy()
```

```python
lista3[1].append(999)
```

```python
lista3
```

```python
lista4
```

```python
print(id(lista3[1]),
id(lista4[1]),
id(lista3),
id(lista4))
```

###solución: Deep copy!

```python
import copy
lista4=copy.deepcopy(lista3)
print(id(lista3[1]),
id(lista4[1]),
id(lista3),
id(lista4))
```
