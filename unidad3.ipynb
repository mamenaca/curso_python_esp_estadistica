```python
id(g),id(copiag)
```

#Ojo, los slices ("pedazos") de arreglo tiene ubicación propia en memoria

```python
id(g[0:1,0:2]),id(copiag[0:1,0:2]),
```

¡Falló usando el copy()! Se puede solucionar haciendo copias de los pedazos

```python
pedazo=copiag[0:1,0:2].copy()
id(g[0:1,0:2]),id(pedazo),
```
