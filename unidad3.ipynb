```python
x=numpy.linspace(-2*numpy.pi, 2*numpy.pi,500)
```

####Método gráfico

```python
pyplot.figure() #crear el lienzo
pyplot.plot(x,y1,"-r") #grafica 1
pyplot.plot(x,y2,"--g") #grafica 2


pyplot.show() #mostrar gráfica
```

###Método redondeo

```python
y2[y1.round(2)==y2.round(2)]
```

###Método de tolerancia

```python
y1[abs(y1-y2)==0]
```

###Método de numpy

```python
numpy.allclose(y1,y2,1e-3) #deben ser igual medida
```
